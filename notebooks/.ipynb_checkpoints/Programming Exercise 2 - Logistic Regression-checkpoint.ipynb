# Logistic Regression

In this part of the exercise, you will build a logistic regression model to predict whether a student gets admitted into a university. Suppose that you are the administrator of a university department and you want to determine each applicant’s chance of admission based on their results on two exams. You have historical data from previous applicants that you can use as a training set for logistic regression. For each training example, you have the applicant’s scores on two exams and the admissions decision.  

Your task is to build a classification model that estimates an applicant’s probability of admission based the scores from those two exams.

```python
%matplotlib inline  

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_1 = pd.read_csv('data/ex2data1.txt', sep=',', header=None, names=['Exam 1', 'Exam 2', 'Admitted'])
data_1.head()
```

## Visualizing the data

Before starting to implement any learning algorithm, it is always good to visualize the data if possible.

```python
def plotData(data_1):
    positive = data_1.loc[data_1['Admitted'] == 1]
    negative = data_1.loc[data_1['Admitted'] == 0]
    
    plt.figure(figsize=(8,6))
    plt.scatter(positive['Exam 1'], positive['Exam 2'], c='black', marker='+', label='Admitted')
    plt.scatter(negative['Exam 1'], negative['Exam 2'], s=20, c='red', marker='x', label='Not Admitted')
    plt.xlabel('Exam 1 score')
    plt.ylabel('Exam 2 score')
    plt.legend(loc=0)
    plt.grid(True)

plotData(data_1)
```

## Implementation

### Warmup exercise: sigmoid function
Before you start with the actual cost function, recall that the logistic regres- sion hypothesis is defined as:
$$
h(x)=g(\theta^Tx),
$$
where function $g$ is the sigmoid function. The sigmoid function is defined as:
$$
g(z) = \frac{1}{1+e^{-z}}
$$

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

For large positive values of $x$, the sigmoid should be close to 1, while for large negative values, the sigmoid should be close to 0. Evaluating sigmoid(0) should give you exactly 0.5.

```python
sigmoid(0)
```

```python
nums = np.arange(-10, 10, step=1)

plt.plot(nums, sigmoid(nums), 'r')
plt.grid(True)
plt.show()
```

### Cost function and gradient
$$ 
J(\theta) = \frac{1}{m}\sum_{i=1}^{m}\big[-y^{(i)}\, log\,( h_\theta\,(x^{(i)}))-(1-y^{(i)})\,log\,(1-h_\theta(x^{(i)}))\big]
$$
Vectorized cost function
$$
J(\theta)=-(\frac{1}{m}((log(g(X\theta))^Ty+(log(1-g(X\theta))^T(1-y)))
$$

```python
def costFunction(theta, X, y):
    m = len(X)
    prediction = sigmoid(np.dot(X, theta))
    cost = -1 / m * (np.log(prediction).T.dot(y) + np.log(1 - prediction).T.dot(1 - y))
    return (cost[0])
```

And the gradient of the cost is a vector of the same length as $\theta$ where the $j^{th}$ element (for $j = 0,1,...,n)$ is defined as follows:
$$
\frac{\partial{J(\theta)}}{\partial{\theta_j}}=\frac{1}{m}\sum_{i=1}^{m}(h_\theta(x^{(i)})-y^{(i)})x_j^{(i)}
$$
Vectorized
$$
\frac{\partial{J(\theta)}}{\partial{\theta_j}}=\frac{1}{m}X^T(g(X\theta)-y)
$$
Note that while this gradient looks identical to the linear regression gradient, the formula is actually different because linear and logistic regression have different definitions of $h_\theta(x)$.

```python
def gradient(theta, X, y):
    m = len(X)
    prediction = sigmoid(X.dot(theta.reshape(-1, 1)))
    grad = (1 / m) * X.T.dot(prediction - y)
    return grad.ravel()
```

```python
X_1 = data_1.iloc[:, :-1].values
m, n = X_1.shape
X_1 = np.insert(X_1, 0, 1, axis=1)
y_1 = data_1.iloc[:, -1:].values
initial_theta = np.zeros((n + 1))
X_1.shape, y_1.shape, initial_theta.shape
```

```python
cost_1 = costFunction(initial_theta, X_1, y_1)

# Cost of initial theta:
print("Cost: \n", cost_1)

grad_1 = gradient(initial_theta, X_1, y_1)

# Gradient of initial theta:
print("Grad: \n", grad_1)
```

### Learning parameters using fminunc

`fminunc` in Octave is a function to do parameter optimizations. In Python, we will use [scipy.optimize module](https://docs.scipy.org/doc/scipy/reference/optimize.html). Notice that by using optimize.minimize, you did not have to write any loops yourself, or set a learning rate like you did for gradient descent. This is all done by optimize.minimize: you only needed to provide a function calculating the cost and the gradient.

```python
from scipy.optimize import minimize

res = minimize(costFunction, initial_theta, args=(X_1, y_1), method=None, jac=gradient, options={'maxiter':400})
res
```

```python
cost = res.fun
print('Cost at theta found by optimize.minimize: {:.3f}'.format(cost))
theta = res.x
print('theta:')
print('\t[{:.3f}, {:.3f}, {:.3f}]'.format(*theta))
```

### Evaluating logistic regression

After learning the parameters, you can use the model to predict whether a particular student will be admitted.

Let's firstly plott the decision boundary using the optimized theta. There is a step by step explanation within the course’s resource on how to plot the decision boundary. The link can be found [here](https://www.coursera.org/learn/machine-learning/resources/fz4AU).

$h_\Theta(x)=g(z)$, where $g$ is the sigmoid function and $z=\Theta^T{x}$      
When $h_\Theta(x) \geq 0.5$, means $\Theta^T{x} > 0$, we predict the class to be "1".    
So the decision boundary is $\Theta_1+\Theta_2x_2+\Theta_3x_3=0$.   
$x_2$ against $x_3$, the boundary line will be the equation $ x_3 = \frac{-(\Theta_1+\Theta_2x_2)}{\Theta_3}$.

```python
def plotDecisionBoundary(data_1):
    boundary_xs = np.array([np.min(X_1[:, 1] - 2), np.max(X_1[:, 1]) + 2])
    boundary_ys = -(theta[0] + theta[1] * boundary_xs) / theta[2]
    plotData(data_1)
    plt.plot(boundary_xs, boundary_ys,'b-',label='Decision Boundary')
    plt.legend()
    
plotDecisionBoundary(data_1)    
```

```python
def predict(theta, X):
    z = np.dot(X, theta)
    prob = sigmoid(z)
    return prob
```

For a student with an Exam 1 score of 45 and an Exam 2 score of 85, you should expect to see an admission probability of 0.776.

```python
X_test_1 = np.array([45, 85])
X_test_1 = np.append(np.ones(1), X_test_1)
pred_prob = predict(theta, X_test_1)
pred_prob
```

Then, we can measure the accuracy of our model. If the predicted probablity is greater or equal to 0.5, we predict the class to be "1", otherwise, "0".

```python
def measureAccuracy(theta, X, y, threshold):
    prob = predict(theta, X)
    pred = prob >= threshold
    accuracy = np.mean(pred == y)
    return accuracy

accuracy = measureAccuracy(theta, X_1, y_1.ravel(), 0.5)
print("Train Accuracy:", accuracy * 100, "%")
```

# Regularized logistic regression

In this part of the exercise, you will implement regularized logistic regression to predict whether microchips from a fabrication plant passes quality assurance (QA). During QA, each microchip goes through various tests to ensure it is functioning correctly.   

Suppose you are the product manager of the factory and you have the test results for some microchips on two different tests. From these two tests, you would like to determine whether the microchips should be accepted or rejected. To help you make the decision, you have a dataset of test results on past microchips, from which you can build a logistic regression model.

```python
data_2 = pd.read_csv('data/ex2data2.txt', sep=',', header=None)
data_2.head()
```

## Visualizing the data

```python
def plotDataReg(data_2):
    positive = data_2.loc[data_2.iloc[:, 2] == 1]
    negative = data_2.loc[data_2.iloc[:, 2] == 0]
    
    plt.figure(figsize=(8, 6))
    plt.scatter(positive.iloc[:, 0], positive.iloc[:, 1], c='black', marker='+', label='y=1')
    plt.scatter(negative.iloc[:, 0], negative.iloc[:, 1], c='red', marker='o', label='y=0')
    plt.xlabel('Microchip Test 1')
    plt.ylabel('Microchip Test 2')
    plt.legend(loc=0)
    plt.grid(True)
    
plotDataReg(data_2)
```

The above figure shows that our dataset cannot be separated into positive and negative examples by a straight-line through the plot. Therefore, a straight- forward application of logistic regression will not perform well on this dataset since logistic regression will only be able to find a linear decision boundary.

```python
X_2 = data_2.iloc[:, :-1]
y_2 = data_2.iloc[:, 2]
X_2.shape, y_2.shape
```

## Feature mapping

One way to fit the data better is to create more features from each data point. We will map the features into all polynomial terms of $x_1$ and $x_2$ up to the sixth power.


\begin{equation}
mapFeature(x)=
\begin{bmatrix}  
1\\x_1\\x_2\\x_1^2\\x_1x_2\\x_2^2\\x_1^3\\ \vdots\\x_1x_2^5\\x_2^6
\end{bmatrix}
\end{equation}

As a result of this mapping, our vector of two features (the scores on two QA tests) has been transformed into a 28-dimensional vector. A logistic regression classifier trained on this higher-dimension feature vector will have a more complex decision boundary and will appear nonlinear when drawn in our 2-dimensional plot.   


While the feature mapping allows us to build a more expressive classifier, it also more susceptible to overfitting. In the next parts of the exercise, you will implement regularized logistic regression to fit the data and also see for yourself how regularization can help combat the overfitting problem.

```python
def mapFeature(X1, X2, degree=6):
    if X1.ndim > 0:
        out = [np.ones(X1.shape[0])]
    else:
        out = [np.ones(1)]

    for i in range(1, degree + 1):
        for j in range(i + 1):
            out.append((X1 ** (i - j)) * (X2 ** j))

    if X1.ndim > 0:
        return np.stack(out, axis=1)
    else:
        return np.array(out)

X_2 = mapFeature(X_2.iloc[:,0], X_2.iloc[:,1], degree=6)
```

## Cost function and gradient

Now you will implement code to compute the cost function and gradient for regularized logistic regression.

Recall that the regularized cost function in logistic regression is:

$$ 
J(\theta) = \frac{1}{m}\sum_{i=1}^{m}\big[-y^{(i)}\, log\,( h_\theta\,(x^{(i)}))-(1-y^{(i)})\,log\,(1-h_\theta(x^{(i)}))\big] + \frac{\lambda}{2m}\sum_{j=1}^{n}\theta_{j}^{2}
$$

Vectorized Cost Function

$$ J(\theta) = -\frac{1}{m}\big((\,log\,(g(X\theta))^Ty+(\,log\,(1-g(X\theta))^T(1-y)\big) + \frac{\lambda}{2m}\sum_{j=1}^{n}\theta_{j}^{2}$$

```python
def costFunctionReg(theta, X, y, reg):
    m = len(X)
    prediction = sigmoid(np.dot(X, theta))
    cost = -1 / m * (np.log(prediction).T.dot(y) + np.log(1 - prediction).T.dot(1 - y))

    # Don't regularize theta_0
    penalty = reg / (2 * m) * np.sum(np.square(theta[1: ]))
    J_theta = cost + penalty
    
    return J_theta
```

The gradient of the cost function is a vector where the $j^{th}$ element is defined as follows (**Notes:** intercept parameter $\theta_{0}$ is not to be regularized):
$$ 
\frac{\partial J(\theta)}{\partial\theta_{j}} = \frac{1}{m}\sum_{i=1}^{m} ( h_\theta (x^{(i)})-y^{(i)})x^{(i)}_{j} + \frac{\lambda}{m}\theta_{j}
$$

Vectorized

$$ 
\frac{\partial J(\theta)}{\partial\theta_{j}} = \frac{1}{m} X^T(g(X\theta)-y) + \frac{\lambda}{m}\theta_{j}
$$

```python
def gradientReg(theta, X, y, reg):
    m = len(X)
    prediction = sigmoid(X.dot(theta.reshape(-1,1)))
    grad = 1 / m * np.dot(X.T, (prediction - y))
    grad[1: ] = grad[1: ] + reg / m * theta[1: ].reshape(-1, 1)
    return grad.ravel()
```

```python
initial_theta = np.zeros(X_2.shape[1])
reg = 1
cost = costFunctionReg(initial_theta, X_2, y_2, reg)
print('The expected cost value is: 0.693')
print('The output cost is: %.5f' % cost)
```

### Learning parameters using fminunc

Again, we will use `fminunc` function to learn the optimal parameter $\theta$.

```python
res_reg = minimize(costFunctionReg, initial_theta, args=(X_2, y_2.reshape(-1, 1), reg), method=None, jac=gradientReg, options={'maxiter':100})
res_reg
```

```python
theta_optimized = res_reg.x

accuracy = measureAccuracy(theta_optimized, X_2, y_2.ravel(), 0.5)
print("Train Accuracy of regularized LR:", accuracy * 100, "%")
```

## Plotting the decision boundary

We plot the non-linear decision boundary by computing the classifier’s predictions on an evenly spaced grid and then and drew a contour plot of where the predictions change from $y = 0$ to $y = 1$

```python
def plotDecisionBoudaryReg(reg, theta, accuracy):
    u = np.linspace(-1, 1.5, 50)
    v = np.linspace(-1, 1.5, 50)
    z = np.zeros((len(u), len(v)))
    
    # Evaluate z = theta*x over the grid
    for i, ui in enumerate(u):
        for j, vj in enumerate(v):
            z[i, j] = np.dot(mapFeature(ui, vj), theta)

    z = z.T  # important to transpose z before calling contour
    # print(z)
    
    plotDataReg(data_2)
    
    # Plot z = 0
    plt.contour(u, v, z, levels=[0], linewidths=2, colors='g')
    plt.title('accuracy = {}%, lambda = {}'.format(np.round(accuracy, 2), reg))
    

plotDecisionBoudaryReg(reg, theta_optimized, accuracy * 100)    
```

In this part of the exercise, you will get to try out different regularization parameters for the dataset to understand how regularization prevents overfitting.

Notice the changes in the decision boundary as you vary $\lambda$. With a small $\lambda$, you should find that the classifier gets almost every training example correct, but draws a very complicated boundary, thus overfitting the data. This is not a good decision boundary: for example, it predicts that a point at $x = (−0.25, 1.5)$ is accepted $(y = 1)$, which seems to be an incorrect decision given the training set.

With a larger $\lambda$, you should see a plot that shows an simpler decision boundary which still separates the positives and negatives fairly well. However, if $\lambda$ is set to too high a value, you will not get a good fit and the decision boundary will not follow the data so well, thus underfitting the data
